## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
from scipy.stats import norm
import scipy.stats as stats


mean_incm = new_df['Income'].mean()
std_dev_incm = new_df['Income'].std()


prob = stats.norm.cdf(2) - stats.norm.cdf(-2)

print(f"Expected value: {mean_incm:.2f}, Standard Deviation: {std_dev_incm:.2f}, Probability: {prob:.2f}")
```

```python
new_df['Income'].plot(kind='kde')
plt.xlabel('Income')
plt.ylabel('Density')
plt.title('Income distribution')
plt.show()

stats.probplot(new_df['Income'], dist='norm', plot=plt)
plt.title('Q-Q plot of Income')
plt.show()
```

```python
mean_Amt_Wines = new_df['Amt_Wines'].mean()
std_dev_Amt_Wines = new_df['Amt_Wines'].std()

prob = stats.norm.cdf(2) - stats.norm.cdf(-2)

print(f"Expected value: {mean_Amt_Wines:.2f}, Standard Deviation: {std_dev_Amt_Wines:.2f}, Probability: {prob:.2f}")
```
